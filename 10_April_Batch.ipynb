```python
import pandas as pd
```

```python
sheet = pd.read_html(page.content)
```

```python
len(sheet)
```

```python
table = sheet[0]
```

```python
table.head()
```

```python
table = table[1:-1]
```

```python
table.shape
```

```python
plt.rcParams['figure.figsize'] = 12, 6
```

```python

plt.bar(table['Country,Other'][:30], table['TotalCases'][:30])
plt.xticks(rotation=90)
plt.show()
```

```python

plt.bar(table['Country,Other'][10:30], table['TotalCases'][10:30])
plt.xticks(rotation=90)
plt.yticks(range(0, 30000, 1000))
plt.grid()
plt.show()
```
